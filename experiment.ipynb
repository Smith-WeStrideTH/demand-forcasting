```python
import os
import json
```

```python
output_dir = 'output'
os.makedirs(output_dir,exist_ok=True)
```

```python
summary = {"abs": 12}
summary_json_path = os.path.join(output_dir, "evaluation_summary.json")

# If SageMaker created a directory at this path (bad ProcessingOutput config), fix it:
if os.path.isdir(summary_json_path):
    # write to a different file name instead
    summary_json_path = os.path.join(output_dir, "evaluation_summary_file.json")
    print("[EVAL] WARNING: evaluation_summary.json is a directory. Writing to:", summary_json_path)

with open(summary_json_path, "w") as f:
    json.dump(summary, f, indent=2)

```
