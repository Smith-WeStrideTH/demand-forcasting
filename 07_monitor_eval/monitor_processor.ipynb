```python
import sagemaker
from sagemaker.sklearn.processing import SKLearnProcessor
import boto3

sess = sagemaker.Session()
role = sagemaker.get_execution_role()
region = boto3.Session().region_name
bucket = sess.default_bucket()

print(region)
```

```python
# ดึง eval_output_s3 ที่ evaluate.py เขียนไว้
%store -r eval_output_s3
print("eval_output_s3:", eval_output_s3)

monitor_base_s3 = f"s3://{bucket}/monitor_retail_demand/"
print("monitor_base_s3:", monitor_base_s3)

```

```python
sk_processor = SKLearnProcessor(
    framework_version="1.2-1",     # หรือเวอร์ชันที่คุณใช้
    role=role,
    instance_type="ml.m5.xlarge",
    instance_count=1,
    base_job_name="ingest-monitor-retail-demand",
    sagemaker_session=sess,
)

sk_processor.run(
    code="ingest_to_athena.py",
    arguments=[
        "--eval-output-s3", eval_output_s3,
        "--monitor-base-s3", monitor_base_s3,
        "--athena-db-name", "monitor_retail_demand",
    ],
    wait=True,
    logs=True,
)
```
